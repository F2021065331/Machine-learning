```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/spam.csv')
df.head()
```

```python
df.info()
```

```python
df['Spam'] = df['Category'].apply(lambda x:1 if x=='spam' else 0)
df = df[['Message', 'Spam']]
df.head()
```

```python
df['Message'] = df['Message'].str.lower()
df.head()
```

```python
# prompt: Create a model to classify emails as spam or not spam based on their
# content.

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

# Ensure x_train contains strings by selecting the 'Message' column AFTER lowercasing
x_train, x_test, y_train, y_test = train_test_split(df['Message'].astype(str), df['Spam'], test_size=0.25)

v = CountVectorizer()
x_train_count = v.fit_transform(x_train) # No need for .values as x_train is already a Series of strings
x_train_count.toarray()[:3]

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train_count, y_train)

emails = [
    'Sounds good, can you please send me the details?',
    'Congratulations! You have won a free gift. Click here to claim.'
]
emails_count = v.transform(emails)
model.predict(emails_count)

x_test_count = v.transform(x_test.astype(str)) # Apply astype(str) to x_test as well for consistency
model.score(x_test_count, y_test)
```

```python
# prompt: Algorithm: Logistic Regression,random forest classifier, adaboosting classifier, knn

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier


# ... (Your existing code for data loading and preprocessing) ...

# Split the data
x_train, x_test, y_train, y_test = train_test_split(df['Message'].astype(str), df['Spam'], test_size=0.25)

# Vectorize the text data
v = CountVectorizer()
x_train_count = v.fit_transform(x_train)
x_test_count = v.transform(x_test)

# Train and evaluate different models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "KNN": KNeighborsClassifier()
}

results = {}
for name, model in models.items():
    model.fit(x_train_count, y_train)
    score = model.score(x_test_count, y_test)
    results[name] = score

# Print the results
for name, score in results.items():
    print(f"{name}: {score}")
```

```python
def predict_spam(email):
    email_count = v.transform([email]) # Transform the new email
    prediction = model.predict(email_count)[0]
    if prediction == 1:
        return "Spam"
    else:
        return "Not Spam"


# Example usage:
emails = [
    'Sounds good, can you please send me the details?',
    'Congratulations! You have won a free gift. Click here to claim.',
    'Hey there! How are you doing? Fancy grabbing a coffee?'
]

for email in emails:
    prediction = predict_spam(email)
    print(f"'{email}' is classified as: {prediction}")
```
